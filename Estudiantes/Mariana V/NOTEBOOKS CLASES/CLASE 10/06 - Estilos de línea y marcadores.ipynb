# Estilos de línea y marcadores

En esta lección vamos a seguir personalizando nuestro gráfico, concretamente las líneas y los marcadores:

```python
import numpy as np
import matplotlib.pyplot as plt

meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
mapeado = range(len(meses))

Mariana = [100,250,450,210,300,550]
Sebastian = [50,75,250,375,250,400]
Santiago = [275,410,200,300,150,200]
```

## Líneas

Tenemos las siguientes propiedades básicas:

- `linewidth (lw)`: Ancho de la línea
- `linestyle (ls)`: Estilo de la línea
- `color`: Color de la línea (número y decimal)
- `alpha`: Opacidad de la línea (de 0 a 1)

```python
# Configuración de las líneas
plt.plot(Mariana, label="Mariana", color="green", ls="-", lw="2")
plt.plot(Sebastian, label="Sebastián", color="red", ls="--", lw="2")
plt.plot(Santiago, label="Santiago", color="blue", ls="-.", lw="2")
plt.xticks(mapeado, meses)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


#Creo lista del eje X, etiquetas, el mapeado le indica al grafico la cantidad de elementos interseptan
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
mapeado = range(len(meses))


#cargo los datos a las lsita para grafiocar cada linea
carlos = [100, 57, 300, 10, 150, 550]
fabian = [10, 20, 30, 10, 15, 55]
mariana = [425, 400, 350, 425, 500, 460]


# Configuración de las líneas y dibujo del grafico eje Y
plt.plot(carlos, label="Carlos", color="red",ls="-", lw="1")


#Aplicó y dibujo el eje x
plt.xticks(mapeado,meses)


#Agregar otra linea al grafico eje y
plt.plot(fabian,label='Fabian', color='#0000ff',ls="--", lw="4")
plt.plot(mariana,label='Mariana', color='green',ls="-.", lw="3")


#opconal si quiero personalizar las leyendas
plt.legend(['Carlos','Fabian','Mariana'])
plt.show()
```

En la [documentación oficial](https://matplotlib.org/api/lines_api.html#matplotlib.lines.Line2D.set_linestyle) puede consultar más opciones de configuración.

## Marcadores

Hacen referencia a los puntos o vértices donde se dibujan los valores:

- `marker`: Tipo de marcador
- `markersize`: Tamaño del marcador
- `markerfacecolor`: Color del marcador (número y decimal)
- `markeredgecolor`: Color del borde (número y decimal)
- `markeredgewidth`: Tamaño del borde

```python
# Configuración de los marcadores
plt.plot(Mariana, label="Mariana", color="red", ls="-", lw="2", 
         marker="o", markersize="10", markeredgewidth="2", markerfacecolor="yellow", markeredgecolor="red")

plt.plot(Sebastian, label="Sebastian", color="#0000ff", ls="--", lw="2",
         marker="*", markersize="15", markeredgewidth="2", markerfacecolor="cyan", markeredgecolor="#0000ff")

plt.plot(Santiago, label="Santiago", color="green", ls="-.", lw="2",
         marker="D", markersize="10", markeredgewidth="2", markerfacecolor="lime", markeredgecolor="green")

plt.xticks(mapeado, meses)
plt.legend()
plt.show()
```
